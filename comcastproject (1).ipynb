```python
#Import the necessary libraries
import numpy as np
import datetime
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import the csv file
df_dataset_comcast = pd.read_csv("Comcast_telecom_complaints_data.csv")
df_comcast_backup = df_dataset_comcast
```

```python
#Provide the trend chart for the number of complaints at monthly and daily granularity levels.
#collecting the data for the same

#extracting the month
df_dataset_comcast['mm'] = pd.to_datetime(df_dataset_comcast['Date_month_year']).dt.month

print(df_dataset_comcast['mm'])
```

```python
#Presenting the trend chart for the number of complaints 
#at monthly granularity levels in the form of axesubplot.

df_dataset_comcast['mm'].value_counts().plot.bar()
```

```python
#Presenting the trend chart for the number of complaints 
#at monthly granularity levels in the form of line chart.

year_and_complaints = pd.DataFrame(df_dataset_comcast['mm'].value_counts())
```

```python
year_and_complaints
```

```python
#convert the month which are index to a column
year_and_complaints = year_and_complaints.reset_index()
year_and_complaints.rename(columns = {'index' :'I'}, inplace = True)
year_and_complaints
```

```python
# arranging the months in serial order
year_and_complaints.sort_values(by = "I",inplace=True)
type(year_and_complaints["I"])
```

```python
year_and_complaints
```

```python
x = year_and_complaints.loc[:,["I"]]
y = year_and_complaints.loc[:,["mm"]]
```

```python
#plot the chart for monthly complaints
plt.figure(figsize = (10,5))
plt.plot(x,y,color = "c", linestyle = "--",linewidth = 3,marker = "o")
plt.xlabel("Months in numbers")
plt.ylabel("No. of Complaints in a month")
plt.title("Comcast Telecom Consumer Complaints")
plt.show()
```

```python
#Extract values for scatter chart for complaints on daily basis
df_daywise_data = df_comcast_backup

#converting to date_month_year column to date type
df_daywise_data['Date_month_year'] = pd.to_datetime(df_dataset_comcast['Date_month_year'])
df_daywise_data.sort_values(by = "Date_month_year", inplace = True)
df_date = df_daywise_data.loc[:,["Date_month_year"]]
df_date
x = []
y = []
c = 0
mm = ""
for i in df_date["Date_month_year"]:
    if(i != mm) :
        dd= str(i)
        x.append(dd[0:10])
        y.append(c)
        c=0                                  
        mm=i
    c = c + 1
```

```python
plt.figure(figsize = (20,10))
plt.scatter(x,y)
plt.xlabel("Day on which complaints were received")
plt.ylabel("No. of Complaints received")
plt.xticks(rotation = 'vertical')
plt.title("Comcast Telecom Consumer Complaints")
plt.show()
```

```python
# creating a new categorial column new status to store OPEN and PENDING as OPEN and
# CLOSED & SOLVED as CLOSED
df_categorical = df_comcast_backup
df_categorical["new_Status"] = df_categorical["Status"].astype("category") 
value_map_new_status = {"Open" :"Open", "Pending" : "Open", "Solved" : "Closed", "Closed" : "Closed"}
df_categorical["new_Status"] = df_categorical["Status"].apply(lambda x : value_map_new_status.get(x))
df_categorical.loc[ : , ["Status","new_Status"]].tail(50)
```

```python
#Extract data for State wise status of complaints
df_categorical.sort_values(by = ["State", "new_Status"], ascending = [True,True]).head(20)
col_list = ["State", "new_Status"]
df_statewise_data = df_categorical[col_list]

#divide the dataframe into 2 dataframes based on the new status OPEN and CLOSED
df1 = pd.DataFrame(df_statewise_data[df_statewise_data["new_Status"] == "Open"])
df2 = pd.DataFrame(df_statewise_data[df_statewise_data["new_Status"] == "Closed"])

# Count the State wise Open rows
y1 = pd.DataFrame(df1["State"].value_counts())

# Count the state wise closed rows
y2 = pd.DataFrame(df2["State"].value_counts())

#In first dataframe rename column "State" as "Opened"
y1 = y1.rename(columns = {"State" : "Opened"})

#In second dataframe rename column "State as "Closed"
y2 = y2.rename(columns = {"State" : "Closed"})
#Merge both the dataframes
if(len(y1) > len(y2)) :
    y3 = y1.join(y2)
else :
    y3 = y2.join(y1)
```

```python
#Fill Nan with 0 for column Opened
y3["Opened"].fillna(0.0,inplace = True)

# Convert the State which are index to a column
y3.reset_index(inplace = True)
```

```python
# rename the column index to State
y3 = y3.rename(columns = {"index" : "State"})
y3
```

```python
#store that data to be plotted on a stacked bar in List
m_state = y3["State"]
m_open = y3["Opened"]
m_close = y3["Closed"]
```

```python
plt.figure(figsize = (20,20))
plt.bar(m_state, m_open,  color = 'r')
plt.bar(m_state, m_close, bottom = m_open, color = 'b')
plt.xticks(rotation = 90)
plt.xlabel("Names of State", fontsize = 18)
plt.ylabel("Status of Complaints", fontsize = 18)
plt.title("State wise status of Complaints", fontsize = 22)
plt.legend(['Open','Closed'], fontsize = 16)

plt.show()
```

```python
# insights into the above chart
#which state has the maximum complaints

print("INSIGHTS BASED ON THE ABOVE CHART")
max_open_complaints = y3["Opened"].max()
l = len(y3)
print("1. State with maximum Open Complaints ")
for i in range(0,43,1) :
    if(m_open[i] == max_open_complaints) :
        print("     ",m_state[i])

#which Stae has maximum closed complaints
max_closed_complaints = y3["Closed"].max()
print("2. State with maximum Closed Complaints ")
for i in range(0,43,1) :
    if(m_close[i] == max_closed_complaints) :
        print("     ",m_state[i])

#which Stae has no Open complaints
print("3. State with no Open Complaints ")
c= 0
for i in range(0,43,1) :
    if(m_open[i] == 0) :
        c = c + 1
        print("     ",m_state[i])
if (c == 0) :
    print("     ","Nil")

#which Stae has no Closed complaints
print("4. State with no Closed Complaints ")
c = 0
for i in range(0,43,1) :
    if(m_close[i] == 0) :
        print("     ",m_state[i])
        c = c + 1
if (c == 0) :
    print("     ","Nil")
```

```python
#Provide the percentage of complaints resolved till date, 
#which were received through the Internet and customer care calls.
complaint_via = df_dataset_comcast["Received Via"].value_counts()
print("==============================")
print("TOTAL COMPLAINTS RECEIVED VIA ")
print("==============================")
print(complaint_via)
print("")
print("")
print("===============================================")
print("CURRENT STATUS OF COMPLAINTS RESOLVED TILL DATE")
print("===============================================")
complaint_closed = df_dataset_comcast["new_Status"].value_counts()
print(complaint_closed)
print("")
print("")
print("============================================================")
print("CURRENT STATUS OF COMPLAINTS RESOLVED TILL DATE RECEIVED VIA")
print("=============================================================")
df_closed_only = pd.DataFrame(df_dataset_comcast[df_dataset_comcast["new_Status"] == "Closed"])
complaint_closed_recd_via = df_closed_only["Received Via"].value_counts()
complaint_closed_recd_via
```

```python
# caclulate percentage of OPEN and CLOSED complaints Received via Customer care call and Internet
complaint_cat = df_dataset_comcast["Received Via"].unique()
print(complaint_cat)
closed_percentage = (100*complaint_closed_recd_via/complaint_via).round(0)
open_percentage = 100 - closed_percentage
print(closed_percentage)
print(open_percentage)
```

```python
#plot the stacked chart as actual data and percentage using SUBPLOT.
plt.figure(figsize = (20,10))
plt.subplot(2,2,1)
plt.bar(complaint_cat, complaint_via-complaint_closed_recd_via,  color = 'g')
plt.bar(complaint_cat, complaint_closed_recd_via, bottom = complaint_via-complaint_closed_recd_via, color = 'y')
plt.xticks(rotation = 90, fontsize = 16)
plt.xlabel("Received Via", fontsize = 18)
plt.ylabel("Status of Complaints Resolved", fontsize = 18)
plt.title("Insight of Compalints (actual data)",fontsize = 22)
plt.legend(['Open','Closed '], fontsize = 16)

plt.subplot(2,2,2)
plt.bar(complaint_cat, open_percentage,  color = 'r')
plt.bar(complaint_cat, closed_percentage, bottom = open_percentage, color = 'b')
plt.xticks(rotation = 90, fontsize = 16)
plt.xlabel("Received Via", fontsize = 18)
plt.ylabel("Status of Complaints Resolved", fontsize = 18)
plt.title("Insight of Compalints (Percentage)",fontsize = 22)
plt.legend(['Open Percentage','Closed Percentage'], fontsize = 16)

plt.show()
```
